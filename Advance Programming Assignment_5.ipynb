<h4 align='justify'><b>Q1. What is the meaning of multiple inheritance?</b></h4>
<h4 align='justify'><b>Multiple Inheritance</b></h4>
<h4 align='justify'> Python provides us the flexibility to inherit multiple base/parent classes in single child class.</h4>

oops2.jpg

<h4><b>Syntax:</b></h4>

```
class Parent1:
  <class-suite>

class Parent2:
  <class-suite>
.
.
.
class ParentN:
  <class-suite>
```
<h4><b>Example:</b></h4>

````python
class Parent1:
  def func1(self):
    print("Hello from Parent1 class")

class Parent2:
  def func2(self):
    print("Hello from Parent2 class")

class Child(Parent1, Parent2):
  def func3(self):
    print("Hello from child class")

test = Child()
test.func1()
test.func2()
test.func3()
````

<h4 align='justify'><b>Q3. What is the concept of composition?</b></h4>
<h4 align='justify'>It is one of the fundamental concepts of Object-Oriented Programming. In this concept, we will describe a class that references to one or more object of other classes as an Instance variable. Here, by using the class name or by creating the object we can access the members of one class inside another class. It enables creating complex types by combining objects of different classes. It means that a class Composite can contain an object of another class Component. This type of relationship is known as <b>Has-A Relation</b></h4>

image.png

<h4 align='justify'>In the above figure Classes are represented as boxes with the class name <b>Composite</b> and <b>Component</b> representing Has-A relation between both of them.</h4>
<h4 align='justify'><b>Syntax:</b>

```
class A:
  # variables of class A
  # methods of class A
  ...
  ...
class B:
  # by using "obj" we can access member's of class A.
  obj = A()

class A.
  obj = A()

  # variables of class B
  # methods of class B

  ...
  ...

```
<h4><b>Example:</b></h4>

````python
class Component:

  # component class constructor
  def __init__(self):
    print("Component class object created...")

  # component class instance method
  def m1(self):  
    print("Component class m1() method executed....")

class Composite:

  # composite class constructor
  def __init__(self):

    # creating object of component class
    self.obj1 = Component()

    print("Composite class object also created...")

  def m2(self):
    print("Composite class m2() method executed...")
    
    # calling m1() method of component class
   
    self.obj1.m1()

# creating object of composite class
obj2 = Composite()


# calling m2() method of composite class
obj2.m2()
````

<h4 align='justify'><b>Q4. What are bound methodsand how do we use them?</b></h4>
<h4><b>BOund Methods</b></h4>
<h4 align='justify'>If a function is an attribute of class and it is accessed via the instance, they are called bound methods. A bound method is one that has 'self' as its first argument. Since these are dependent on the instance of classes, these are also known as instance methods.</h4>
<h4 align='justify'><b></b>Need for these bound methods</b></h4>
<h4 align='justify'>The methods inside the classes would take at least one argument. To make them zero-argument methods, 'decorators' has to be used. Different instance of a class have different values associated with them.</h4>
<h4 align='justify'><b>Example:</b><h4>

````python
class Sample(object):

  # static variable for object number
  objectNo = 0

  def __init__(self, name1):

    # variable to hold name
    self.name = name1

    # Increment static variable for each object
    Sample.objectNo = Sample.objectNo + 1

    # each object's unique number that can be consider as ID
    self.objNumber = Sample.objectNo

  def myFunc(self):
    print(f"My name is {self.name}, from objcet {self.objNumber}")

  def alterIt(self, newName):
    self.name = newName

  def myfunc2():
    print("I am not a bound method !!!")

# creating first instance of class Sample
samp1 = Sample("A")
samp1.myFunc()

# unhide the line below to see the error 
# samp1.myFunc2() #---------> error line

# creating second instance of class sample
samp2 = Sample("B")
samp2.myFunc()
samp2.alterIt("C")
samp2.myFunc()
samp1.myFunc()
             
````
