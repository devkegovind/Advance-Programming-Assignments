<h4 align='justify'><b>Walrus Assignment Operator</b></h4>

*   One of the biggest change in this version is the introduction of <b>assignment operator.</b> They are represented using the symbol. <b>:=</b>

*   They are generally used to assign and return values in the same expression.

For example in Python 3.7
```
foo = "something"
print(something)
```
In Python 3.8 we could write this as
```
print(foo:="something")
```

````python
print(foo:="something")
````

*   The <b>Walrus operator</b> can also be used inside `if` or `while` blocks that calculate a value to test a particular condition.

```
# check whether the first element of the list is 24

while (foo:=a[0]) == 24:
  process (foo)

```

<h4 align='justify'><b>Position Only Parameter</b></h4>

*   The new function parameter syntax`/` 
which indicates for some function parameters to be specified <b>Positionally</b> They can't be used as <b>keyword assignments</b>

````python
# a, b are positional only arguments

def foo(a, b, /, c, d):
  print(a, b, c, d)

# you can use these as function calls
foo(8, 9, 10, d=11)

# These are invalid calls
foo(a = 8,9, 10, 11)
foo(8, b = 9, 10, 11)
````

*   One of my favourite addition is the support of `=` operator to `f-strings` specifiers. This can be really helpful in debugging and self-documenting expressions.

<h4 align='justify'><b>3. What is the difference between a shallow copy and deep copy?</b></h4>
<h4 align='justify'><b>Shallow Copy</b>

*   A shallow copy creates a new object which stores the references of the original elements.
*   So, a shallow copy doesn't create a copy of nested objects, instead it just copies the reference of nested objects. This means, a copy process does not recurse or create copies of nested objects itself.
Example:

````python
# Create a copy using shallow copy
import copy

old_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
new_list = copy.copy(old_list)

print(f"Old List {old_list}")
print(f"New List {new_list}")



# Adding [4, 4, 4] to old_list, using shallow copy
import copy

old_list = [[1, 1, 1], [2, 2, 2], [3, 3, 3]]
new_list = copy.copy(old_list)

old_list.append([4, 4, 4])

print(f"Old List {old_list}")
print(f"New List {new_list}")

# Adding new nested object using Shallow copy

old_list = [[1, 1, 1], [2, 2, 2], [3, 3, 3]]
new_list = copy.copy(old_list)

old_list[1][1] = "devke"

print(f"Old List {old_list}")
print(f"New List {new_list}")
````

<h4 align='justify'>We add the new list i.e.[4, 4, 4] into old_list. This new sublist was not copied in new_list.</h4>


<h4 align='justify'><b>Deep Copy</b></h4>

*   A Deep copy creates a new object and recursively adds the copies of nested objects present in the original elements. 
*   Let's continue with example we are going to create deep copy using `deepcopy()` function present in `copy` module. The deep copy creates independent copy of original object and all its nested objects.

Example:



````python
# Copying a list using deepcopy()
import copy

old_list = [[1, 1, 1], [2, 2, 2], [3, 3, 3]]
new_list = copy.deepcopy(old_list)

print(f"Old List {old_list}")
print(f"New List {new_list}")

# Adding a new nested object in the list using deep copy

import copy

old_list = [[1, 1, 1], [2, 2, 2], [3, 3, 3]]
new_list = copy.deepcopy(old_list)

old_list[1][0] = "Govind"

print(f"Old List {old_list}")
print(f"New List {new_list}")
````

<h4 align='justify'>In the above program, when we assign a new value to old_list, we can see only the old_list is modified. This means, both the old_list and the new_list are independent. This is because the old_list was recursively copied, which is true for all its nested objects.</h4>

<h4 align='justify'><b>5. What is generator comprehension?</b></h4>
<h4 align='justify'>A generator comprehension is <b>a single-line specification for defining a generator in Python.</b> It is absolutely essential to learn this syntax in order to write simple and readable code.</h4>
<h4 align='justify'><b>Example:</b></h4>

````python
# find out even number and make it square
gen_exp = (x**2 for x in range(11) if x % 2 == 0)

for x in gen_exp:
  print(x)
````
