<h4 align = 'justify'><b>
Q6. In standard expressions, what is the difference between positive look-ahead and negative look-
ahead?</b></h4>
<h4 align = 'justify'>Positive look ahead is an assertion continuing the search and extending the string e.g. pattern r'abc(?=[A-Z])' Here after 'abc', ? is extending the search and says that in the remaining string, first identify the next.</h4>
<h4 align = 'justify'>Example:</h4>

```python
import re
pat = r'abc(?=[A-Z])'
text = "abcABCEF"
regobj = re.compile(pat)
matchobj = regobj.findall(text)
print(f"Positive Lookahead {matchobj}")

# Negative look head is also an assertion to exclude certain patterns.
# e.g. pattern = r'abc(?!abc)', means identify a substring containing
# 'abc' which is not followed by another 'abc'
# Example of Neagtive Lookahead

import re
pat1 = r'abc(?!abc)'
text1 = "aeiouabcabc"
regobj1 = re.compile(pat1)
matchobj1 = regobj1.findall(text)
print(f"Negative Look Ahead {matchobj1}")
```
