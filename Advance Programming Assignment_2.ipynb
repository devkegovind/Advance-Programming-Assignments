<h4 align='justify'><b>Q1. What is the relationship between classes and modules?</b></h4>
<h4 align='justify'>In Python, Modules are collections of methods and constants. They cannot generate instance. Classes may generate instance(objects), and have per-instance state(instance variables.)</h4>

<h4 align='justify'><b>Q2. How do you make instances and classes?</b></h4>
<h4 align = 'justify'> To create instance of a class, you call the class using class name and pass in whatever arguments its __init__ method accepts.</h4> 
<b><h4>Instance of a class</b></h4>
<h4 align='justify'>When you create an object, you are creating an instance of a class, therefore "instantiating" a class. The new operator requires a single, posfix argument: a call to a constructor. The name of the constructor provides the name of the class to instantiate.</h4>

<h4><b>Create a class</b></h4>
<h4 align='justify'>A class is a code template for creating objects. Objects have member variables and have behaviour associated with them. In Python a class is created by the keyword class.</h4>

<h4><b>Q3. Where and how should be class attributes created?</h4></b>
<h4 align='justify'>To define a class attribute, you place it outside of the __init__() method. Use class_name. class_attribute or object_name. class_attribute to access the value of the class_attribute.</h4>

<h4 align='justify'><b>Q4. Where and how are instance attributes created?</b></h4>
<h4 align='justify'>An Instance attribute is a Python variable belonging to only one object. It is only accessible in the scope of the object and it is defined inside the constructor function of a class. For example, __init__(self).</h4>

<h4 align='justify'><b>Q5. What does the term &quot;self&quot; in a Python class mean?</b></h4>
<h4 align='justify'>self represents the instance ofthe class. By using the "self" we can access the attributes and methods of the class in python. It binds the attributes with the given attributes. The reason you need to use self. is because Python does not use the @ syntax to refer to instance attributes.</h4>

<h4 align='justify'><b>Q6. How does a Python class handle operator overloading?</b></h4>
<h4 align='justify'>The operator overlaoding in Python means <b>provide extended meaning beyond their predefined operational meaning.</b> Such as, we use the "+" operator for adding two integers as well as joining two string or merging two list. We can achieve this as the "+" operator is overloaded by the "int" class and "str" class.</h4>

<h4 align='justify'><b>Q7. When do you consider allowing operator overloading of your classes?</b></h4>
<h4 align='justify'>When one or both operands are of a user-defined class or structure type, operator overloading makes it easier to specify user-defined implementation for such operations. This makes user-defined types more similar to the basic primitive data types in terms of behaviour.</h4>

<h4 align='justify'><b>Q8. What is the most popular form of operator overloading?</b></h4>
<h4 align='justify'>A very popular and convenient example is the <b>Addition (+) operator.</b> Just think how the '+' operator operates on two numbers and the same operator operates on two strings. It performs "Addition" on numbers whereas it performs "Concatenation" on string.</h4>

<h4 align='justify'><b>Q9. What are the two most important concepts to grasp in order to comprehend Python OOP code?</b></h4>
<h4 align='justify'>Both <b>inheritance and polymorphism</b> are fundamental concepts of object oriented programming. These concepts help us to create code that can be extended and easily maintainable.</h4>
