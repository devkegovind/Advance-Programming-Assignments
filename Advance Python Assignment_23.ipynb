Q5. Compare and contrast the three types of array multiplication between two numpy arrays: dot
product, outer product, and regular multiplication of two numpy arrays.</b></h4>
<h4 align='justify'>

```python
import numpy as np

a1 = np.array([[1, 2, 3], [4, 5, 6], [6, 7, 8]])
a2 = np.array([[10, 20, 30], [40, 50, 60], [60, 70, 80]])

# suppose above are the two arrays of shape 3 * 3

print(a1)
print()
print(a2)
```

```python
# standard multiplication
a1 * a2
```

<h4 align='justify'>When we do standard multiplication in that case values of the some indexes in the array will get multiplicatins like in above example. a1(i, j) * a2(i, j), dot product</h4>

```python
np.dot(a1, a2)
```

<h4 align='justify'>In case of dot product vector, multiplications will take place between row of first array and column of second array respectively. Like first row of array at will be multiply by column value of array a2 one by one and then added.</h4>

```python
np.outer(a1, a2)
```

<h4 align='justify'>In outer multiplication every element of first array a1 will be multiply by every of other a2 such the number of columns will be equal to the number of element in another array a2.</h4>
