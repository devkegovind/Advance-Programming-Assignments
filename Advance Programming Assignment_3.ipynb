<h4 align='justify'><b>1. What is the concept of an abstract superclass?</b></h4>

<h4 align='justify'>A class is called an Abstract class if it contains one or more abstract methods. An abstract method is a method is a method that is declared, but contains no implementation. Abstract classes may not be instantiated, and its abstract methods must be implemented by its subclasses.</h4>

<h4 align='justify'><b>3. Why does a class need to manually call a superclass&#39;s __init__ method?</b></h4>

<h4 align='justify'><b>4. How can you augment, instead of completely replacing, an inherited method?</b></h4>

<h4 align='justify'>A more sophisticted way to augment an inherited method involves forwarding. <b>Message forwarding</b> allows you to augment an inherited method in such a way that it can perform its inherited action and some new action.</h4>

<h4 align='justify'><b>5. How is the local scope of a class different from that of a function?</b></h4>

<h4 align='justify'>For example, if you assign a value to a name inside a function, then name will have a local Python scope. In contrast, if you asign a value to a name outside of all functions--say, at the top level of a module-- then that name will have a global Python scope.</h4>
