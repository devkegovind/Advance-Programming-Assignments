<h4 align = 'justify'><b>Q4. In Python 3.X, what are the key differences between text-mode and binary-mode files?</b></h4>
<h4 align = 'justify'>Files opened in binary mode(appending 'b' to the mode aragument) return contents as bytes objects without any decoding. In text mode(the default, or when 't' is appended to the mode argument), the contents of the file are returned as strings, the bytes having been first decoded using a platform-depedent encoding or using the specified encoding if given.</h4>

<h4 align = 'justify'><b>Q5. How can you interpret a Unicode text file containing text encoded in a different encoding than
your platform&#39;s default?</b></h4>
<h4 align = 'justify'>While reading file, we can specify the encoding such as(utf16, utf32), so the file will be decoded according to that encoding and result are returned to the user.</h4>

```python
text = "this is sample text file for understanding encoding"
f = open('sample.txt', 'w', encoding = 'utf16')
f.write(text)
f.close()

f = open('sample.txt', 'r', encoding = 'utf16')
print(f.read())
```
