<h4 align='justify'><b>Q1. What is the difference between __getattr__ and __getattribute__?</b></h4>
<h4 align='justify'>Normally, if we want to access the attribute of a class, we access it with instance of the class like instance.attribute_name. These attributes are already present in the class. Where as,to access the attribute of class which are not defined in the class, we use <b>getattr()</b> method.</h4>
<h4 align='justify'>But if the attribute does exist, <b>getattr</b> won't be invoked.</h4>

```python
class Sample:
  def __init__(self, address):
    self.address = address

  def __getattr__(self, name):
    return name.lower()

s = Sample('Pune')
print(s.address) # It is present in the class.
print(s.ABCD) # It is not present in the class.
```

<h4 align='justify'>getattribute will look for every attribute, doesn't matter if the attribute exists or not.</h4>

```python
class Dummy():

  def __getattribute__(self,attr):
    return 'New Value'

d = Dummy()
d.value = "Old value"
print(d.value)
```

<h4 align='justify'>In properties, We can bind getter, setter functions with an attribute name, using the built-in property function. In descriptor, We can bind getter, setter(add deleter) functions into a seperate class. We then assign an object of this class to the attribute name.</h4>
