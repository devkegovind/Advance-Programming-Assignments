<h4 align='justify'>The try block lets you test a block of code for errors. The except block lets handle the error.The else block lets you execute when there is no error.</h4>

```

  # block of code
```

````python
try:
  num = int(input("enter a number: "))
  assert num%2 == 0
except:
  print("Not an even number.")
else:
  reciprocal = 1/num
  print(reciprocal)
````
