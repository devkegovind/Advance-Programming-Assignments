<h4 align='justify'><b>Q11. What is the one improvement that must be made to a function in order for it to become a generator function in the Python 
language?</b></h4>
<h4 align='justify'>We can use "yield" keyword instead of "return" to make function a generator.</h4>

```python
def even_nos(n):
  for i in range(n):
    if i % 2 == 0:
      yield i

for i in even_nos(10):
  print(i)
```

<h4 align='justify'><b>
Q12. Identify at least one benefit of generators.</b></h4>
<h4 align='justify'>The main advantage here is that instead of having to compute an entire series of values up front, the generator computes one value and then suspends its activity awaiting the next instruction. This feature is known as state suspension.</h4>
<h4 align='justify'>Generators are memory efficient.</h4>
