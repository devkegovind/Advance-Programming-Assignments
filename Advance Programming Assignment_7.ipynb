<h4 align='justify'><b>Q1. What is the purpose of the try statement?</b></h4>
<h4 align='justify'>The try statement works as follows:</h4>

- <h4 align='justify'>First, try clause (the statement(s) between the try and except keywords) is executed.</h4>
- <h4 align='justify'>If no exception occurs, the except clause is skipped and execution of the try statement if finished.</h4>

- <h4 align='justify'>If an exception occurs during exception of the try clause, the rest of the clause is skipped.</h4>

<h4 align='justify'>The different Try/Except Variations. So far we'he used a try/except and even a try/except/except, but this is only two-thirds of the story. There are two other optional segements to a try block: else and finally. Both of these optional blocks will come after try and the except.</h4>

<h4 align='justify'><b>Q3. What is the purpose of the raise statement?</b></h4>
<h4 align='justify'>Python raise Keyword is used to <b> raise exceptions or errors.</b> The raise keyword raises an error and stops the control flow of the program. It is used to bring up the current exception in an exception handler so that it can be handled further up the call stack.</h4>

<h4 align='justify'><b>Q4. What does the assert statement do, and what other statement is it like?</b></h4>
<h4 align='justify'>The assert keyword is used when debugging code. The assert keyword <b>lets you test if a condition in your code returns True, if not, the program will raise an AssertionError.</b> You can write a message to be written if the code returns False.</h4>

<h4 align='justify'><b>Q5. What is the purpose of the with/as argument, and what other statement is it like?</b></h4>
<h4 align='justify'>The with statement is a replacement for commonly used try/finally error-handling statements. A common example of using the with statement is opening a file. To open and write to a file in Python.</h4>
<h4 align='justify'><b>Example:</b>

```
with open("test2.txt", "w") as f:
  f.write("Govind\n")
  f.write("Devke\n")
  print(f.read())

```
